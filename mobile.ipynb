```python
df["model_brand"] = df["model"].str.split(" ").apply(lambda x: x[0])
```

```python
value_count=df["model_brand"].value_counts()
to_remove = value_count[value_count<=4].index
df = df[~df.model_brand.isin(to_remove)]
```

```python
def fetch_processor(text):
    if text == 'snapdragon' or text == 'helio' or text == 'dimensity' or text =='exynos' or text =='bionic'or text =='unisoc'or text =='tiger'or text =='google':
        return text
    else:
         return 'Other Processor'
```

```python
df['processor_brand']=df['processor_brand'].apply(fetch_processor)
```

```python
df1 = df.pop('price')
df['price']=df1
```

```python
df.drop('num_cores',axis=1,inplace=True)
df.drop('fast_charging_available',axis=1,inplace=True)
```

```python
df.info()
```

```python
X = df.drop(columns=['price'])
y = np.log(df['price'])
```

```python
X2 = pd.get_dummies(X, columns=['model_brand', 'os', 'processor_brand', 'model'],drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X2,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_absolute_error
from xgboost import XGBRegressor
```

```python
step2 = GradientBoostingRegressor(n_estimators=600)
step2.fit(X_train,y_train)

y_pred = step2.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
