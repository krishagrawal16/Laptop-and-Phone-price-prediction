```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df['Ram'] = df['Ram'].str.replace('GB','')
df['Weight'] = df['Weight'].str.replace('kg','')
```

```python
df.head()
```

```python
df['Ram'] = df['Ram'].astype('int32')
df['Weight'] = df['Weight'].astype('float32')
```

```python
df.info()
```

```python
df["Cpu_Speed"] = df["Cpu"].str.split(" ").apply(lambda x: x[-1]).str.replace("GHz", "")
df["Cpu_Vender"] = df["Cpu"].str.split(" ").apply(lambda x: x[0])
df["Cpu_Type"] = df["Cpu"].str.split(" ").apply(lambda x: x[1:4] if x[1] == "Celeron" and "Pentium" and "Xeon" else (x[1:3] if (x[1] == "Core" or x[0] == "AMD") else x[0]) )
df["Cpu_Type"] = df["Cpu_Type"].apply(lambda x: ' '.join(x))
df["Cpu_Type"]
df.head()
```

```python
df['Cpu_Type'].unique()
```

```python
import seaborn as sns
sns.distplot(df['Price'])
```

```python
split_mem = df['Memory'].str.split(' ', expand=True)
df['Storage_Type'] = split_mem[1]
df['Memory'] = split_mem[0]
df["Memory"].unique()
```

```python
df.head()
```

```python
df['Memory'].unique()
```

```python
df["Memory"]=df["Memory"].str.replace("GB","")
df["Memory"]=df["Memory"].str.replace("TB","")
df["Memory"]=df["Memory"].astype('float')
df.rename(columns={'Memory':'Memory (GB or TB)'}, inplace=True)

df['Memory (GB or TB)'] = df['Memory (GB or TB)'].apply(lambda x: 1024 if x==1 else x)
df['Memory (GB or TB)'] = df['Memory (GB or TB)'].apply(lambda x: 2048 if x==2 else x)
df.rename(columns={'Memory (GB or TB)':'Storage (GB)'}, inplace=True)
df.head()
```

```python
df.info()
```

```python
gpu_distribution_list = df["Gpu"].str.split(" ")
df["Gpu_Vender"] = df["Gpu"].str.split(" ").apply(lambda x: x[0])
df["Gpu_Type"] = df["Gpu"].str.split(" ").apply(lambda x: x[1:])
df["Gpu_Type"] = df["Gpu_Type"].apply(lambda x: ' '.join(x))
df.head()
```

```python
df['Touchscreen'] = df['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
df['Ips'] = df['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
df.head()
```

```python
def cat_os(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Others/No OS/Linux'

df['OpSys'] = df['OpSys'].apply(cat_os)
```

```python
df.head()
```

```python
df.info()
```

```python
new = df['ScreenResolution'].str.split('x',n=1,expand=True)
```

```python
df['X_res'] = new[0]
df['Y_res'] = new[1]
```

```python
df.sample(5)
```

```python
df['X_res'] = df['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
```

```python
df['X_res'] = df['X_res'].astype('int')
df['Y_res'] = df['Y_res'].astype('int')
```

```python
df.info()
```

```python
df['ppi'] = (((df['X_res']**2) + (df['Y_res']**2))**0.5/df['Inches']).astype('float')
```

```python
df.drop(columns=['X_res','Y_res','ScreenResolution','Gpu','Cpu'],inplace=True)
```

```python
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
X_train.info()
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,5,9,10,11,8])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,5,9,10,11,8])
],remainder='passthrough')

step2 = Ridge(alpha=10)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
